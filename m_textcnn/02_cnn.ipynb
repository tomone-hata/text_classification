```python
from cnn import CNN
import pickle
import numpy as np
import pandas as pd
```

## 学習データ・テストデータの読み込み

```python
x_train, y_train = None, None
with open('train_data.pickle', 'rb') as rb:
    x_train, y_train = pickle.load(rb)

x_test, y_test = None, None
with open('test_data.pickle', 'rb') as rb:
    x_test, y_test = pickle.load(rb)
    
cost_weight = {}
with open('cost_weight.pickle', 'rb') as rb:
    cost_weight = pickle.load(rb)
```

## 学習モデルの読み込み

```python
cnn = CNN(vocab_size=143585, embedding_dim=256, sequence_length=512, \
                  num_filters=[3, 3, 3], filter_size=[256, 256, 256],drop=0.2, num_classes= 16, \
                  learning_rate=0.01, beta_1=0.9, beta_2=0.999, epsilon=1e-8, decay=0,\
                  batch_size=32, epochs=5, verbose=1, class_weight=cost_weight)
```

## 学習の開始

```python
train_size = int(len(x_train)  * 0.9)
x_valid, y_valid = x_train[train_size:], y_train[train_size:]
x_train, y_train = x_train[:train_size], y_train[:train_size]
print('Train size        : %i' % len(y_train))
print('Validation size:  %i' % len(y_valid))
cnn.train(x_train, y_train, x_valid, y_valid)
```

## 検証

```python
cnn.test(x_test, y_test)
```

```python
result = np.zeros((16,16))
y_actually = np.argmax(y_test, axis=1)
y_predict = np.argmax(cnn.model.predict(x_test), axis=1)
y_len = y_actually.shape[0]
print(y_len)
for i in range(y_len):
    result[y_actually[i], y_predict[i]] += 1
result = result.astype(int)
```
